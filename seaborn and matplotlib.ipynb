```python
flights = sns.load_dataset('flights')
flights.head()
```

### **Wide Format**
In the **wide format**, each row represents a single observation, and each variable is spread across multiple columns. This format is useful for easily comparing different variables for each observation. 

To convert from long to wide, use `pivot()` or `pivot_table()` in pandas.

### **Long Format**
In the **long format**, each row represents a single observation-variable pair, with a new column holding the value. This format is typically used for easier data manipulation, statistical analysis, and visualisations.

To convert from wide to long, use `melt()` in pandas.


```python
#long to wide format
flights_wide = flights.pivot(index='year', columns='month', values='passengers')
flights_wide.head()
```

## Titanic task

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
missing_values = titanic.isnull().sum()
missing_values
```

```python
duplicates = titanic.duplicated().sum()
duplicates
```

```python
titanic['age'].fillna(titanic['age'].median(), inplace=True)
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)
titanic['fare'].fillna(titanic['fare'].median(), inplace=True)
titanic.head()
```

```python
#average fare and age for each class
average_class = titanic.groupby('pclass')[['fare', 'age']].mean()
average_class
```

```python
#new column added showing the fare as a percentage of total
total_fare = titanic['fare'].sum()
titanic['fare percentage'] = (titanic['fare'] / (total_fare)) * 100
titanic.head()
```
